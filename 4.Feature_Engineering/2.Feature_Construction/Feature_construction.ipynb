<h2 style="color:gold; text-align:left; font-size:200%; font-weight:bold;">
1) Feature Construction</h2>

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression

#Data ingestion 
df=sns.load_dataset('titanic')
df= df[['pclass', 'age', 'sibsp', 'parch', 'survived']]

#Droping the missing values 
df=df.dropna()
df.info()
#<---------------------------------------------------------------------------------------------->
#Train-test split 
x=df.iloc[:, :-1]
y=df.iloc[:, -1]
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state=42)

#<---------------------------------------------------------------------------------------------->
#Before Fearture construction:Model Training > fiting > prediction > accuracy score> cross validation
lr =LogisticRegression()
lr.fit(x_train, y_train) #traing 
y_pred=lr.predict(x_test) #Prediction 
print("lr model sccuracy:", accuracy_score(y_test, y_pred)) 
#cross validation 
print('lr cross val score',np.mean(cross_val_score(lr, x , y, scoring='accuracy', cv=25)))

#<---------------------------------------------------------------------------------------------->

#AAFTER FEATURE CONSTRCUTION :: 
#-----------creating new cols by combining (sibsp , parch & passenger itself)
x['family_size']=x['sibsp']+x['parch']+1
#now create a function that define family type(alone, small, large)
def my_func(num):
    if num==1:
        return 0 #alone 
    elif num>1 and num<=4:
        return 1 #small family 
    else:
        return 2 #large family 
#now apply the function 
x['family_type']=x['family_size'].apply(my_func)
print(x.sample(2))

#now make remove (sibsp, parch, family_size) cols from input
x.drop(columns=['sibsp', 'parch', 'family_size'], inplace=True)
print(x.sample(3))

#Now check the croos_val_score
print('lr cross val score',np.mean(cross_val_score(lr, x , y, scoring='accuracy', cv=25)))
```
